```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import expm

# Sample matrix A (this will determine the behavior of the curve)
A = np.array([[-2/100, 3/100], 
              [2/100,  -3/100]])

# Sample vector c (initial condition)
c = np.array([[1],
              [-1]])

# Time values
t_values = np.linspace(-100, 100, 400)

# Compute x(t) = e^{At}c for each t
def x_t(A, t, c):
    return np.dot(expm(A * t), c)

# Get the x(t) values for each time point
x_values_1 = [x_t(A, t, c)[0][0] for t in t_values]
x_values_2 = [x_t(A, t, c)[1][0] for t in t_values]

# Plot
plt.figure(figsize=(8, 6))

plt.plot(t_values, x_values_1, label='Solution 1', color='blue')
plt.plot(t_values, x_values_2, label='Solution 2', color='orange')

plt.title(r"$x(t) = e^{At}c$")
plt.xlabel("t")
plt.ylabel("x(t)")
plt.legend()
plt.grid(True)
plt.show()
```
